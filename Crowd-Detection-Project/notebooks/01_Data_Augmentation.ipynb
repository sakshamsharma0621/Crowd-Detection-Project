```python
import os
from PIL import Image

# Input folder
input_folder = r"C:/Users/profr/Saksham/dataset/not_crowded"

# Output folder
output_folder = r"C:/Users/profr/Saksham/dataset/temp"

# Create temp folder if not exists
os.makedirs(output_folder, exist_ok=True)

# Loop through all files
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tif', '.tiff')):
        
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path).convert("RGB")   # ensure RGB mode

        # Rotate +30 degrees with white background
        rotated_plus = img.rotate(30, expand=True, fillcolor=(255, 255, 255))
        rotated_plus.save(os.path.join(output_folder, f"{filename}_plus30.jpg"))

        # Rotate -30 degrees with white background
        rotated_minus = img.rotate(-30, expand=True, fillcolor=(255, 255, 255))
        rotated_minus.save(os.path.join(output_folder, f"{filename}_minus30.jpg"))

print("Rotation complete! Images saved in temp folder.")
```

```python
import os
import numpy as np
from PIL import Image, ImageEnhance, ImageOps

# -----------------------------
# Path Configuration
# -----------------------------
source_folder = r"C:/Users/profr/Saksham/dataset/not_crowded"
output_folder = r"C:/Users/profr/Saksham/dataset/temp"

os.makedirs(output_folder, exist_ok=True)

# -----------------------------
# Function: Add Gaussian Noise
# -----------------------------
def add_gaussian_noise(image, mean=0, std=20):
    np_img = np.array(image)
    noise = np.random.normal(mean, std, np_img.shape)
    noisy_img = np.clip(np_img + noise, 0, 255).astype(np.uint8)
    return Image.fromarray(noisy_img)

# -----------------------------
# Loop through all images
# -----------------------------
for filename in os.listdir(source_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.tif', '.tiff')):
        
        img_path = os.path.join(source_folder, filename)
        img = Image.open(img_path).convert("RGB")

        # 1️⃣ Contrast Adjustment (+2 levels)
        contrast_enhancer = ImageEnhance.Contrast(img)
        img_contrast = contrast_enhancer.enhance(2.0)   # increase contrast
        img_contrast.save(os.path.join(output_folder, f"{filename}_contrast2.jpg"))

        # 2️⃣ Reduce Saturation (Color reduction)
        color_enhancer = ImageEnhance.Color(img)
        img_saturation = color_enhancer.enhance(0.4)    # reduce color saturation
        img_saturation.save(os.path.join(output_folder, f"{filename}_saturation_low.jpg"))

        # 3️⃣ Hue Adjustment
        img_hsv = img.convert('HSV')
        h, s, v = img_hsv.split()
        np_h = (np.array(h) + 40) % 255    # shift hue by +40 value
        img_hue = Image.merge("HSV", (Image.fromarray(np_h), s, v)).convert("RGB")
        img_hue.save(os.path.join(output_folder, f"{filename}_hue_shift.jpg"))

        # 4️⃣ Gaussian Noise
        img_noise = add_gaussian_noise(img)
        img_noise.save(os.path.join(output_folder, f"{filename}_gaussian_noise.jpg"))

print("All augmented images saved successfully in temp folder!")
```
